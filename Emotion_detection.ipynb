```python
import cv2
from deepface import DeepFace

# Initialize webcam
cap = cv2.VideoCapture(0)  
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=6, minSize=(60, 60))

    for (x, y, w, h) in faces:
        # Extract and resize face ROI for better emotion detection
        face_roi = frame[y:y+h, x:x+w]
        face_resized = cv2.resize(face_roi, (150, 150))  # Resize for DeepFace

        try:
            # Analyze emotion with better face detection
            result = DeepFace.analyze(
                face_resized, 
                actions=['emotion'], 
                enforce_detection=True,  # Force real face detection
                detector_backend='mtcnn'  # Use MTCNN for better face recognition
            )

            # Get detected emotion and confidence
            dominant_emotion = result[0]['dominant_emotion']
            confidence = result[0]['emotion'][dominant_emotion]

            # Draw bounding box
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

            # Display detected emotion and confidence
            text = f"{dominant_emotion} ({confidence:.1f}%)"
            cv2.putText(frame, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        except Exception as e:
            print("DeepFace Error:", e)

    # Show frame
    cv2.imshow("Real-time Emotion Detection", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
